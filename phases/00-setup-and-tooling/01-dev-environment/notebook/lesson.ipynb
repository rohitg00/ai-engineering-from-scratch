## Layer 1 + 2: System & Package Manager

````python
print(f"Python: {sys.version}")
print(f"uv:     {shutil.which('uv')}")
print(f"git:    {shutil.which('git')}")
print(f"node:   {shutil.which('node')}")
print(f"cargo:  {shutil.which('cargo')}")
````

## Layer 3: Language Runtimes — Python version check

````python
assert sys.version_info >= (3, 11), f"Need Python 3.11+, got {sys.version}"
print(f"Python {sys.version_info.major}.{sys.version_info.minor} — OK")
````

## Layer 4: AI/ML Libraries — NumPy tensor operation

````python
import numpy as np

print(f"NumPy {np.__version__}")

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(f"a: {a}")
print(f"b: {b}")
print(f"dot(a, b) = {np.dot(a, b)}")  # 4 + 10 + 18 = 32
print(f"a @ b = {a @ b}")              # same, using matmul operator

assert np.dot(a, b) == 32
````

## GPU Check (optional)

````python
try:
    import torch
    print(f"PyTorch {torch.__version__}")
    cuda = torch.cuda.is_available()
    mps = torch.backends.mps.is_available()
    print(f"CUDA: {cuda}")
    print(f"MPS (Apple Silicon): {mps}")
    if cuda:
        print(f"GPU: {torch.cuda.get_device_name(0)}")
    elif mps:
        print("GPU: Apple Silicon MPS")
    else:
        print("No GPU — CPU-only. That's fine for most lessons.")
except ImportError:
    print("PyTorch not installed. Install with: uv pip install torch")
    print("No GPU — CPU-only. That's fine for most lessons.")
````
