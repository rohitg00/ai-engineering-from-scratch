# Сколько занимает list comprehension?» / «А numpy?» / «Сравним напрямую

```python
t_list_1 = %timeit -o [random.random() for _ in range(100_000)]
t_list_1.average
```

```python
t_list_2 = %timeit -o np.random.random(100_000)
t_list_2.average
```

## Сравниваем среднее время работы двух способов

```python
speedup = t_list_1.average / t_list_2.average
speedup
```

Результат - в 8 раз быстрее работает способ с помощью numpy. Если я правильно понимаю, потому, что бибилотека написана на C.
