```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_loaded = pd.read_csv("./sales.csv")
df_loaded
```

```python
df_loaded.describe()
```

```python
df_loaded["sales"].mean()
```

```python
mean_sales = df_loaded["sales"].mean()
total_profit = df_loaded["profit"].sum()
peak_month = df_loaded.loc[df_loaded["sales"].idxmax(), "month"]

print(f"Средние продажи за месяц: {mean_sales:.1f}")
print(f"Суммарная прибыль за год: {total_profit}")
print(f"Пиковый месяц продаж: {peak_month}")
```

```python
plt.figure(figsize=(10, 6))

plt.plot(df_loaded["month"], df_loaded["sales"], marker="o", label="Продажи")
plt.plot(df_loaded["month"], df_loaded["expenses"], marker="s", label="Расходы")
plt.plot(df_loaded["month"], df_loaded["profit"], marker="^", label="Прибыль")

plt.title("Динамика по месяцам")
plt.xlabel("Месяц")
plt.ylabel("Сумма")
plt.xticks(df_loaded["month"])
plt.legend()
plt.grid(True, alpha=0.3)

plt.show()
```
